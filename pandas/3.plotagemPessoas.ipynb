## 3.2 Insight B

```python
filtro4=uniao['tipo_envolvido']=='Condutor'
```

```python
condutor=uniao.loc[filtro4]
```

##3.3 Insight C

```python
uniao.groupby(['fase_dia']).size().sort_values(ascending=False).plot.bar(figsize=(10,5))
```
